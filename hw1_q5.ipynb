## Import Data

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Auto.csv')
```

```python
data = data.replace('?', np.nan) # For the question marks
data = data.dropna() # Drop all rows with at least one missing value
```

```python
data.head()
```

```python
# (b)
quantitative_features = ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "year"]
range = data[quantitative_features].astype(float).max() - data[quantitative_features].astype(float).min()
```

```python
range
```

```python
# (c)
data[quantitative_features].astype(float).mean()
```

```python
data[quantitative_features].astype(float).std()
```

## Now let us remove the 10th through 85th observations

```python
new_data = data.drop(data.index[9:85])
```

```python
new_data
```

```python
range = new_data[quantitative_features].astype(float).max() - new_data[quantitative_features].astype(float).min()
range
```

```python
new_data[quantitative_features].astype(float).mean()
```

```python
new_data[quantitative_features].astype(float).std()
```
