## Importing libraries

## Reading and loading data (a)

```python
college = pd.read_csv("College.csv")
college.head()
```

## (b)

```python
college2 = pd.read_csv("College.csv", index_col=0)
college2.head()
```

```python
college3 = college.rename({"Unnamed: 0": "College"}, axis=1)
college3 = college3.set_index("College")
college3.head()
```

```python
college = college3
```

## Describe the data (c)

```python
college.describe()
```

## Make a scatter matrix (d)

```python
pd.plotting.scatter_matrix(college[["Top10perc", "Apps", "Enroll"]], figsize=(10,10), alpha=0.3)
```

## Create a boxplot (e)

```python
college.boxplot(column='Outstate', by='Private', figsize=(8, 6))
```

## (f)

```python
college['Elite'] = pd.cut(college['Top10perc'], [0, 50, 100], labels=['No', 'Yes'])
college["Elite"].value_counts()
```

```python
college.boxplot(column='Outstate', by='Elite', figsize=(8, 6))
```

## Histograms (g)

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0, 0].hist(college['Top10perc'], bins=5)
axes[0, 0].set_title('Bins = 5')

axes[0, 1].hist(college['Top10perc'], bins=10)
axes[0, 1].set_title('Bins = 10')

axes[1, 0].hist(college['Top10perc'], bins=20)
axes[1, 0].set_title('Bins = 20')

axes[1, 1].hist(college['Top10perc'], bins=50)
axes[1, 1].set_title('Bins = 50')

plt.show()
```
