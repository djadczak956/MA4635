```python
# For Q3
def f(n):
  return 1 - (1 - 1/n) ** n
```

```python
for i in [5, 100, 10000, 100000]:
  print(f'f({i}) = {f(i)}')
```

## Plot that displays the phenomenon. (g)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 100000)
y = [f(n) for n in x]

plt.plot(x, y)
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title('Plot of f(n) from 1 to 100000')
plt.show()
```
